```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
for i in df:
    if df[i].dtype == "O":
        df[i]=le.fit_transform(df[i])
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
x = df.drop('class',axis = 1)
y = df['class']
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=10000)
```

```python
from sklearn.feature_selection import RFE
```

```python
d = {}
for i in range(1,len(x.columns)):
    rfe = RFE(model,n_features_to_select=i)
    x_data=pd.DataFrame(rfe.fit_transform(x,y))
    from sklearn.model_selection import KFold
    kf = KFold(n_splits=8)
    from sklearn.model_selection import cross_val_score
    cv_sc= list(cross_val_score(model,x_data,y,cv=kf))
    train,test = list(kf.split(x_data,y))[cv_sc.index(max(cv_sc))]
    x_train = x_data.iloc[train]
    y_train = y.iloc[train]
    x_test = x_data.iloc[test]
    y_test = y.iloc[test]
    model.fit(x_train,y_train)
    scr = model.score(x_test,y_test)
    d[i]=scr
```

```python
accu_k= max(d,key=d.get)
```

```python
print('Best accuracy achieved was from ',accu_k,"Features\naccuracy = ",d[accu_k])
```
